```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import accuracy_score, mean_squared_error

# Using the KMeans from sklearn
from sklearn.cluster import AgglomerativeClustering
```

### 1.Load the file “zoo.data” and look at the info and first five rows. The first column denotes the animal name and the last one specifies a high-level class for the corresponding animal. 

```python
data = pd.read_csv('zoo.csv')
```

###  2. Find out the unique number of high level class.

```python
data['class_type'].unique()
```

### 3. Use the 16-intermediate feature and perform an agglomerative clustering. 

```python
x = data.drop(['class_type','animal_name'],axis=1)
y = data['class_type']
```

```python
agglo = AgglomerativeClustering(n_clusters=7)
predicted_values = agglo.fit_predict(x)
```

### 4. Compute the mean squared error by comparing the actual class and predicted highlevel class.

```python
# Accuracy Score
print("Accuracy Score")
print(accuracy_score(predicted_values, y))

# Mean Square Error value
print("Mean Squared Error value")
print(mean_squared_error(predicted_values, y))
```
