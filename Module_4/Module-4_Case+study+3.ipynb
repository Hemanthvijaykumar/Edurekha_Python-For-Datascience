```python
data = pd.DataFrame({'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
'female': [0, 1, 1, 0, 1],
'age': [42, 52, 36, 24, 73],
'preTestScore': [4, 24, 31, 2, 3],
'postTestScore': [25, 94, 57, 62, 70]})
```

```python
data
```

```python
plt.scatter(x=data['preTestScore'], y=data['postTestScore'],s=data['age'])
plt.title('scatter plot',fontsize=14)
plt.xlabel('x-axis',fontsize=14)
plt.ylabel('y-axis',fontsize=14)
plt.show()
```

```python
# Question 10: Draw a Scatterplot from the data in question 9 of preTestScore and postTestScore with the size = 300 and the color determined by sex
```

```python
plt.scatter(x=data['preTestScore'], y=data['postTestScore'],s=300,c=data['female'])
plt.title('scatter plot',fontsize=14)
plt.xlabel('x-axis',fontsize=14)
plt.ylabel('y-axis',fontsize=14)
plt.show()
```
